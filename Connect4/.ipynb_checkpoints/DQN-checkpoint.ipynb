# Deep Q Networks: Tetris Agents

The DQN architecture employs neural networks for the function approximation of states to state-action values (Q-values). In this Tetris DQN, the state input is the game image in greyscale. The neural network learns to extract the visual features of the game image in the convolutional layers, learning a feature-vector representation of the state space. The network can use then map the state space to Q values using dense neural network layers (also known as linear/feed-forward layers or multi-layered perceptrons).
